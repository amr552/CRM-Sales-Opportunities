```python
df_sales_pipeline['deal_stage'] = df_sales_pipeline['deal_stage'].replace(['Engaging', 'Prospecting'], 'Open')
df_sales_pipeline['deal_stage'].value_counts()
```

```python
df_sales_pipeline['engage_date']=pd.to_datetime(df_sales_pipeline['engage_date'])
df_sales_pipeline['close_date']=pd.to_datetime(df_sales_pipeline['close_date'])
df_sales_pipeline.drop(['close_sate'],axis=1,inplace=True)
df_sales_pipeline.info()
```

```python
df_sales_pipeline['account'] = df_sales_pipeline['account'].fillna('Unknown Account')
df_sales_pipeline['account']
```

```python
df_master=df_sales_pipeline.merge(df_accounts,on='account',how='left')
df_master.head()
```

```python
df_master=df_master.merge(df_product,on='product',how='left')
df_master=df_master.merge(df_sales_teams,on='sales_agent',how='left')
df_master.head()
```

```python
df_master.info()
```

```python
df_master['win_flag'] = df_master['deal_stage'].apply(lambda x: 1 if x == 'Won' else 0)
```

```python
df_master['deal_age'] = (df_master['close_date'] - df_master['engage_date']).dt.days
```

```python
df_master['year'] = df_master['engage_date'].dt.year.astype('Int64')
df_master['quarter'] = df_master['engage_date'].dt.quarter.astype('Int64')
df_master['month'] = df_master['engage_date'].dt.month.astype('Int64')
df_master.head()
```

```python
#Won revenue
df_master['won_revenue'] = df_master.apply(
    lambda row: row['close_value'] if row['deal_stage']=='Won' else 0,
    axis=1
)
```

```python
df_master['lost_revenue'] = df_master.apply(
    lambda row: row['close_value'] if row['deal_stage']=='Lost' else 0,
    axis=1
)
```
